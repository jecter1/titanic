## 3. Get some info from train data

```python
def print_surv_info(data_part):
    survived_cnt = len(data_part[data_part.Survived == 1])
    not_survived_cnt = len(data_part[data_part.Survived == 0])
    all_cnt = survived_cnt + not_survived_cnt
    perc = round(survived_cnt / all_cnt * 100, 2)
    print(f'Survived: {survived_cnt}/{all_cnt} ({perc}%)')
```

## 8. Fill missing values

### 8.1. Common functions

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

def make_dataset(data, features_cols, target_col, out=True):
    X = data[features_cols].to_numpy()
    y = data[target_col].to_numpy().ravel()
    
    scaler = StandardScaler()
    X = scaler.fit_transform(X)
    
    X_test = X[np.isnan(y)]
    X_train = X[~np.isnan(y)]
    y_train = y[~np.isnan(y)]
    
    if out:
        print(f'Train size: {X_train.shape[0]}')
        print(f'Test size: {X_test.shape[0]}')
        print(f'Num features: {X_train.shape[1]}')
    
    return X_train, y_train, X_test
```

### 8.2. Fill Fare

```python
data.loc[data.Fare.isna()]
```

```python
similar_passengers = data.loc[(data.Pclass == 3) & (data.TicketCount == 1) & (data.Deck == 0) & (data.Age > 40)]
similar_passengers.Fare
```

```python
NUM_BINS = 15
plt.figure(figsize=(12, 8))
sns.histplot(similar_passengers.Fare, bins=NUM_BINS, kde=True, color='g')
plt.title('Fare of similar passengers')
plt.show()
```

```python
fare_fill = similar_passengers.Fare.median()
fare_fill
```

```python
data.Fare.fillna(fare_fill, inplace=True)
```

### 8.3. Fill Age using regressor

```python
data.columns
```

```python
data.info()
```

```python
features_cols = list(data.columns)

to_remove = ('PassengerId', 'Survived', 'Name', 'Ticket', 'Cabin', 'Age')
for rmv in to_remove:
    features_cols.remove(rmv)

target_col = ['Age']
```

```python
X_train, y_train, X_test = make_dataset(data, features_cols, target_col)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import GradientBoostingRegressor

params = {
    'n_estimators': (100,),
    'learning_rate': (0.1, 0.05, 0.02, 0.01),
    'max_depth': [i for i in range(2, 11)],
    'random_state': (42,)
}
cv = GridSearchCV(GradientBoostingRegressor(), params, scoring='neg_mean_squared_error', n_jobs=-1, cv=2)
cv.fit(X_train, y_train)
print(f'Best score: {abs(cv.best_score_)}')
print(f'Best params:')
for param_name, param_val in cv.best_params_.items():
    print(f'\t{param_name}: {param_val}')
```

```python
model = cv.best_estimator_
y_test_pred = model.predict(X_test)
print(f'min predicted value = {y_test_pred.min()}')
print(f'max predicted value = {y_test_pred.max()}')
target_name = target_col[0]
data.loc[data[target_name].isna(), target_name] = y_test_pred
```

## 9. Predict Survived

```python
featires_cols = ['Sex', 'Pclass', 'Deck', 'Fare', 'Embarked_C']
target_col = ['Survived']

X_train, y_train, X_test = make_dataset(data, features_cols, target_col)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

params = {
    'n_estimators': (100,),
    'learning_rate': (0.1, 0.05, 0.02, 0.01),
    'max_depth': [i for i in range(2, 11)],
    'random_state': (42,)
}
cv = GridSearchCV(GradientBoostingClassifier(), params, scoring='accuracy', n_jobs=-1, cv=2)
cv.fit(X_train, y_train)
print(f'Best score: {cv.best_score_}')
print(f'Best params:')
for param_name, param_val in cv.best_params_.items():
    print(f'\t{param_name}: {param_val}')
```

```python
model = cv.best_estimator_
y_test_pred = model.predict(X_test)
```
